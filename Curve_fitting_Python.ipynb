```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,10,11)
```

I actually used a random number generator to add noise, but here is what my Quiz used for data. 

```python
y = [3.5, 2.7, 2.1, 1.8, 1.3, 1.1, 0.7, 0.6, 0.5, 0.4, 0.1]
```

```python
plt.plot(t,np.round(y,1),'*')
plt.xlabel('time')
plt.ylabel('torque')
plt.title('torque versus time- definitely not straight')
plt.grid(True)
```

```python
plt.plot(np.log(t), np.log(y), "o")
plt.xlabel("time (sec)")
plt.ylabel("torque (ft-lbf)")
plt.title("torque versus time, power is definitely not straight either- divide by zero is concerning (this model is WRONG)")
plt.grid(True)
```

```python
plt.plot(t, np.log(y), "o")
plt.xlabel("time (sec)")
plt.ylabel("torque (ft-lbf)")
plt.title("torque versus time, exponential is much better")
plt.grid(True)
```

```python
p = np.polyfit(t,np.log(y),1)
p
```

```python
m = p[0]
print('m = ', m)
b = np.exp(p[1])
print('b = ', b)
```

```python
plt.plot(t, np.round(y, 1),'*', t, b*np.exp(m*t))
plt.xlabel('time')
plt.ylabel('torque')
plt.title('torque versus time')
plt.grid(True)
```
